```python
import requests
import re
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = 'https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=14450;type=tournament'

response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the anchor tags <a> which contain hyperlinks
anchor_tags = soup.find_all('a')

# Initialize an empty list to store the extracted links
links_with_series = []

# Regular expression pattern to match links containing '/series'
pattern = re.compile(r'/series')

# Iterate through the anchor tags and extract the 'href' attribute
for anchor_tag in anchor_tags:
    link = anchor_tag.get('href')
    if link and pattern.search(link):  # Check if 'href' attribute exists and matches the pattern
        full_link = urljoin('https://www.espncricinfo.com/', link)
        links_with_series.append(full_link)
        all_links = links_with_series[:-3]

# Print each link on a new line
for link in all_links:
    print(link)
```

```python
def batting_data(url):
    # Extracting HTML from URL
    page = requests.get(url).text
    soup = BeautifulSoup(page, 'html.parser')

    # Find team innings information
    pattern = r'<span class="ds-text-tight-xs">(.*?) Innings<\/span>'
    matches = re.findall(pattern, page)
    
    if len(matches) >= 2:
        teams_playing = matches[:2]
    else:
        print("Pattern not found in the HTML snippet.")
        return None
    
    # Find batting tables
    tables = soup.find_all('table', class_='ci-scorecard-table')

    # Extract data from batting tables
    def extract_innings_data(table, match_info, team_innings):
        data = []
        for row in table.find_all('tr'):
            columns = row.find_all(['th', 'td'])
            row_data = [column.get_text(strip=True) for column in columns]
            data.append(row_data)

        columns = data[0]
        df = pd.DataFrame(data[1:], columns=columns, index=None)
        df = df.dropna()
        df = df.rename(columns={"BATTING": "batsmanName", "": "dismissal"})
        df['match'] = match_info
        df['teamInnings'] = team_innings
        df['battingPos'] = range(1, len(df['batsmanName']) + 1)
        df = df.drop(columns='M')
        df = df[['match', 'teamInnings', 'battingPos', 'batsmanName', 'dismissal', 'R', 'B', '4s', '6s', 'SR']]
        return df

    df_list = []
    for i in range(2):  # Iterate through both teams
        df_list.append(extract_innings_data(tables[i], f"{teams_playing[0]} Vs {teams_playing[1]}", teams_playing[i]))

    result_df = pd.concat(df_list, ignore_index=True)
    return result_df



# Process each URL
final_result = pd.DataFrame()
for url in all_links:
    df = batting_data(url)
    if df is not None:
        final_result = pd.concat([final_result, df], ignore_index=True)
```

```python
def extract_bowling_data(url):
    # Extracting HTML from URL
    page = requests.get(url).text
    soup = BeautifulSoup(page, 'html.parser')

    # Find team innings information
    pattern = r'<span class="ds-text-tight-xs">(.*?) Innings<\/span>'
    matches = re.findall(pattern, page)
    
    if len(matches) >= 2:
        teams_playing = matches[:2]
    else:
        print("Pattern not found in the HTML snippet.")
        return None
    
    # Find bowling tables
    target_tables = [table for table in soup.find_all('table') if any('BOWLING' in header.get_text().upper() for header in table.find_all('th'))]

    # Extract data from bowling tables
    def extract_innings_data(table, match_info, team_innings):
        data = []
        for row in table.find_all('tr'):
            columns = row.find_all(['th', 'td'])
            row_data = [column.get_text(strip=True) for column in columns]
            data.append(row_data)

        columns = data[0]
        df = pd.DataFrame(data[1:], columns=columns, index=None)
        df = df.dropna()
        df = df.rename(columns={"BOWLING": "bowlerName", "O": "overs", "M": "maiden", "R": "runs", "W": "wickets", "ECON": "economy", "WD": "wides", "NB": "noBalls"})
        df['match'] = match_info
        df['bowlingTeam'] = team_innings
        df = df[['match', 'bowlingTeam', 'bowlerName', 'overs', 'maiden', 'runs', 'wickets', 'economy', '0s', '4s', '6s', 'wides', 'noBalls']]
        return df

    df_list = []
    for i in range(2):  # Iterate through both teams
        df_list.append(extract_innings_data(target_tables[i], f"{teams_playing[0]} Vs {teams_playing[1]}", teams_playing[1 - i]))  # Switch the indices

    result_df = pd.concat(df_list, ignore_index=True)
    return result_df

# Process each URL
final_result1 = pd.DataFrame()
for url in all_links:
    df = extract_bowling_data(url)
    if df is not None:
        final_result1 = pd.concat([final_result1, df], ignore_index=True)
```

```python
import requests
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
pd.set_option('display.max_rows', None)
```

```python
#new
# Initialize a list to store HTML content
html_pages = []

# Iterate through the URLs and retrieve HTML content
for url in all_links:
    response = requests.get(url)
    if response.status_code == 200:
        html_pages.append(response.content.decode('utf-8'))
    else:
        print("Failed to retrieve page:", url)
```

```python
#new

# Combine HTML content of all pages
combined_html = '\n'.join(html_pages)

# Create a single BeautifulSoup object from the combined HTML
soup = BeautifulSoup(combined_html, 'html.parser')
```

```python
#new

tables = soup.find_all('table', class_='ci-scorecard-table')
unique_links = set()  # Using a set to store unique links

for table in tables:
    anchor_tags = table.find_all('a')
    links = ['https://www.espncricinfo.com' + tag['href'] for tag in anchor_tags]
    unique_links.update(links)  # Add links to the set

# Convert the set back to a list, if needed
player_links = list(unique_links)
```

```python
#new
# Initialize an empty list to store the HTML content of each page
player_html_pages = []

for link in player_links:
    response = requests.get(link)
    if response.status_code == 200:
        player_html_pages.append(response.content.decode('utf-8'))  # Store HTML content
    else:
        print(f"Failed to retrieve data from {link}")

# Combine HTML content of all pages
player_combined_html = '\n'.join(player_html_pages)

# Create a single BeautifulSoup object from the combined HTML
player_soup = BeautifulSoup(player_combined_html, 'html.parser')
```

```python
player_profiles = player_soup.find_all('div',class_ ='ds-p-0')
# Initialize lists for player names and team names
player_names = []
team_names = []

# Iterate through the range of indices from 0 to the length of player_profiles
for i in range(len(player_profiles)):
    # Find team name within the profile, or set to empty string if not found
    team_name_element = player_profiles[i].find('span',class_ ='ds-text-comfortable-s')
    team_name = team_name_element.get_text() if team_name_element else ""
    
    # Find player name within the profile, or set to empty string if not found
    player_name_element = player_profiles[i].find('h1')
    player_name = player_name_element.get_text() if player_name_element else ""

    # Append the names to the respective lists
    player_names.append(player_name)
    team_names.append(team_name)
```

```python
p_names = [item for item in player_names if item != '']
t_names = [item for item in team_names if item != '']
```

```python
#new

player_details = player_soup.find_all('div', class_='ds-grid-cols-2')
# Create empty lists to store extracted details
batting_style_list = []
bowling_style_list = []
playing_role_list = []

# Define the strings you want to find
strings_to_find = ['Batting Style', 'Bowling Style', 'Playing Role']

# Iterate through each 'div' element in player_details
for detail in player_details:
    # Create a temporary BeautifulSoup object for the current detail element
    temp_soup = BeautifulSoup(str(detail), 'html.parser')

    # Find all 'p' elements within the temporary soup object
    p_elements = temp_soup.find_all('p')

    # Initialize variables for details
    batting_style = None
    bowling_style = None
    playing_role = None

    # Iterate through the found 'p' elements
    for p_element in p_elements:
        text = p_element.get_text(strip=True)
        if 'Batting Style' in text:
            batting_style = p_element.find_next('p').get_text(strip=True)
        elif 'Bowling Style' in text:
            bowling_style = p_element.find_next('p').get_text(strip=True)
        elif 'Playing Role' in text:
            playing_role = p_element.find_next('p').get_text(strip=True)

    # Append details to respective lists
    batting_style_list.append(batting_style)
    bowling_style_list.append(bowling_style)
    playing_role_list.append(playing_role)

# Create a DataFrame from the lists
data = {
    'Batting Style': batting_style_list,
    'Bowling Style': bowling_style_list,
    'Playing Role': playing_role_list
}
details_df = pd.DataFrame(data)
```

```python
# Initialize an empty list to store lists of descriptions
all_descriptions = []

# Iterate through each HTML page
for html_page in player_html_pages:
    # Create BeautifulSoup object for the current HTML page
    player_soup = BeautifulSoup(html_page, 'html.parser')

    # Find 'div' elements with class 'ci-player-bio-content'
    desc = player_soup.find_all('div', class_='ci-player-bio-content')

    # Initialize an empty list for descriptions in this page
    descriptions = []

    # Populate the descriptions list for this page
    for d in desc:
        description_text = d.text.strip() if d and d.text.strip() else None
        descriptions.append(description_text)

    # Append the descriptions for this page to the overall list
    all_descriptions.append(descriptions)
```

```python
# Add 'Description' column to the existing DataFrame
details_df['Description'] = all_descriptions
```

```python
# Convert 'Description' column to string type
details_df['Description'] = details_df['Description'].astype(str)
# Function to remove square brackets and double quotes
def remove_brackets_and_quotes(description):
    description = re.sub(r'\[(.*?)\]', r'\1', description)  # Remove square brackets
    description = description.replace('"', '')  # Remove double quotes
    description = description.replace("'", '')  # Remove single quotes
    return description

# Apply the function to the entire 'Description' column
details_df['Description'] = details_df['Description'].apply(remove_brackets_and_quotes)
```

```python
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

image_urls = []  # List to store image URLs

# Iterate through the player names and URLs
for player_name, url in zip(player_names, player_links):
    driver.get(url)

    try:
        # Explicitly wait for the image element to be present
        wait = WebDriverWait(driver, 200)  # Wait up to 200 seconds (adjust as needed)

        # Find all image elements on the page
        image_elements = driver.find_elements(By.TAG_NAME, 'img')

        # Iterate through image elements to find matching alt attribute
        for image_element in image_elements:
            alt_text = image_element.get_attribute('alt').lower()
            if player_name.lower() in alt_text:
                # Scroll to trigger lazy loading
                driver.execute_script("arguments[0].scrollIntoView();", image_element)
                time.sleep(7)  # Wait for images to load

                # Get the src attribute (image URL)
                image_url = image_element.get_attribute('src')
                image_urls.append(image_url)  # Add the image URL to the list
                break  # Exit loop after finding the first matching image URL
        else:
            image_urls.append("")  # Add an empty string if no matching image found
    except Exception as e:
        image_urls.append("")  # Add an empty string if there's an error
        print("Error:", e)
        # Quit the driver after all players have been processed
driver.quit()
```
